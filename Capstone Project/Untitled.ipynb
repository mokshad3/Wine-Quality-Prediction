# Model Training

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

import pickle

import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.simplefilter(action='ignore', category=ConvergenceWarning)
```

```python
#target variable
y = df.quality
#input feature
X = df.drop('quality', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.2,
                                                   random_state=1234)
```

# Creating Pipeline

```python
make_pipeline(StandardScaler(), RandomForestRegressor(random_state=123))
```

```python
make_pipeline(StandardScaler(), GradientBoostingRegressor(random_state=123))
```

```python
make_pipeline(StandardScaler(), Lasso(random_state=123))
make_pipeline(StandardScaler(), Ridge(random_state=123))
make_pipeline(StandardScaler(), ElasticNet(random_state=123))
```

```python
# Creating dictionary for pipelines
pipelines = {
    'lasso' : make_pipeline(StandardScaler(), Lasso(random_state=123)),
    'ridge' : make_pipeline(StandardScaler(), Ridge(random_state=123)),
    'rf' : make_pipeline(StandardScaler(), RandomForestRegressor(random_state=123)),
    'gb' : make_pipeline(StandardScaler(), GradientBoostingRegressor(random_state=123)),
    'enet': make_pipeline(StandardScaler(), ElasticNet(random_state=123))
}
```

# Hyperparameters

```python
lasso_hyperparameters = { 
    'lasso__alpha' : [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10] 
}

ridge_hyperparameters = { 
    'ridge__alpha': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10]  
}
enet_hyperparameters = { 
    'elasticnet__alpha': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10],                        
    'elasticnet__l1_ratio' : [0.1, 0.3, 0.5, 0.7, 0.9]  
}
```

```python
rf_hyperparameters = { 
    'randomforestregressor__n_estimators' : [100],
    'randomforestregressor__max_features': ['auto', 'sqrt', 0.33],
}
```

```python
gb_hyperparameters = { 
    'gradientboostingregressor__n_estimators': [100, 200],
    'gradientboostingregressor__learning_rate' : [0.05, 0.1, 0.2],
    'gradientboostingregressor__max_depth': [1, 3, 5]
}
```

```python
# Creating Dictionary for hyperparameters
hyperparameters = {
    'lasso' : lasso_hyperparameters,
    'ridge' : ridge_hyperparameters,
    'enet' : enet_hyperparameters,
    'rf' : rf_hyperparameters,
    'gb' : gb_hyperparameters
}
```

```python
fitted_models = {}


for name, pipeline in pipelines.items():
    model = GridSearchCV(pipeline, hyperparameters[name], cv=10, n_jobs=-1)
    
    
    model.fit(X_train, y_train)
    
    
    fitted_models[name] = model
    
   
    print(name, 'has been fitted.')
```

```python
for key, value in fitted_models.items():
    print( key, type(value) )
```

```python
from sklearn.exceptions import NotFittedError

for name, model in fitted_models.items():
    try:
        pred = model.predict(X_test)
        print(name, 'has been fitted.')
    except NotFittedError as e:
        print(repr(e))
```

```python
for name, model in fitted_models.items():
    print( name, model.best_score_ )
```

```python
for name, model in fitted_models.items():
    pred = model.predict(X_test)
    print( name )
    print( '--------' )
    print( 'R^2:', r2_score(y_test, pred ))
    print( 'MAE:', mean_absolute_error(y_test, pred))
    print()
```

```python
rf_pred = fitted_models['rf'].predict(X_test)
plt.scatter(rf_pred, y_test)
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```
